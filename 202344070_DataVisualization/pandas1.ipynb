```python
import pandas as pd

data = {'Name': ['John', 'Alice', 'Bob', 'Eve', 'Charlie'], 
        'Age': [25, 30, 22, 35, 28], 
        'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'], 
        'Salary': [50000, 55000, 40000, 70000, 48000]}

df = pd.DataFrame(data)
print(df)
```

이 경우는 loc랑 큰 차이가 없음

```python
# iloc은 숫자 위치 (0에서 시작)
second_row = df.iloc[1]
print(second_row)
```

iloc는 뒤에거 포함 안함 근데 그냥 loc는 포함함

```python
# df.loc[a:b]에서 b는 포함
# df.loc은 언제나 인덱스 레이블을 기준
print(df)
print("-"*100)
subset = df.loc[0:2, ['Name', 'Age']]
print(subset)
```

df는 5개이기에 5개 전부가 나오고 이제 저건 주어진 값을 비교
마스킹이란 일종의 필터 해당하는걸 가져오는거임

```python
# 마스크
df['Age'] > 25
```

필터링 해서 해당하는 것만 출력

```python
# 불리언 인덱싱
filtered_data = df[df['Age'] > 25]
print(filtered_data)
```

```python
import pandas as pd

# 샘플 데이터 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45],
    'City': ['New York', 'London', 'Paris', 'Tokyo', 'Seoul'],
    'Score': [85, 92, 78, 95, 88]
}

df = pd.DataFrame(data, index=['p1', 'p2', 'p3', 'p4', 'p5'])

df
```

```python
# 인덱스 이름(레이블)이 p2인 사람의 'City' 정보를 가져오려면? .loc를 사용하여 코드를 한 줄로 작성
value = df.loc['p2','City']
print(value)
```

```python
# 데이터프레임의 첫 번째 행부터 세 번째 행까지만 선택하려면? .iloc를 사용하여 코드를 한 줄로 작성
df3 = df.iloc[0:3]
print(df3)
```

```python
# Score가 90점 이상인 데이터들만 필터링하여 전체 컬럼을 출력하려면?
df[df['Score'] >= 90]
```

```python
# 나이가 35세 이상인 사람들의 Name과 Score 컬럼을 출력하려면?
df[df['Age'] >= 35][['Name','Score']]
```
