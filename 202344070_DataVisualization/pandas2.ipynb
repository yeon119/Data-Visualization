```python
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Score': [85, 90, 95, 80]}
df = pd.DataFrame(data)
df
```

sort_values가 따로 정해지지 않았으니 age 기준으로 오름차순으로 정렬(기본값)

```python
sorted_df = df.sort_values(by='Age')
print(sorted_df)
```

ascending이 false 이므로 내림차순

```python
sorted_df = df.sort_values(by='Age',ascending=False)
print(sorted_df)
```

ignore_index(True)라 하면 인덱스를 정렬에 포함하지 않음이라는거 위랑 다른점을 비교하면 보임 
주의할점은 원래 인덱스가 뭔지 모르기떄문에 조심해야하ㅁ

```python
sorted_df = df.sort_values(by='Age',ascending=False, ignore_index=True)
print(sorted_df)
```

```python
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Score': [85, 90, 95, 80]}
df = pd.DataFrame(data)
df
```

age가 같은경우 다른 조건으로 score로 비교해서 순서를 정함

```python
# 'Age' 값이 같은 경우, 'Score'로 sorting
sorted_df = df.sort_values(by=['Age', 'Score'])
print(sorted_df)
```

원본 데이터에는 none이지만 표시될때는 nan으로 표시됨 missing data는 시각화랑 연결됨 
데이타프레임을 만들때 데이터는 csv파일에 있음 

```python
import pandas as pd
data_with_nan = {"Name": ["Alice", "Bob", "Charlie", "David"],"Age": [28, 22, None, 22]}
df_nan = pd.DataFrame(data_with_nan)
df_nan
```

sorting을 할떄 결측치가 있으면 그 값에대한 기본값은 마지막으로 감

```python
# default: 결측치는 마지막으로

# sorted_df = df_nan.sort_values(by="Age", na_position="first")
sorted_df = df_nan.sort_values(by="Age")
print(sorted_df)
```

```python
# 인덱스 정렬
```

```python
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Score': [85, 90, 95, 80]}
df = pd.DataFrame(data)
df
```

index를 기준으로 정렬 기본값은 ascending

```python
# default는 ascending 순서
df_sorted_by_index = df.sort_index()
print(df_sorted_by_index)
```

```python
df_sorted_by_index_desc = df.sort_index(ascending=False)
print(df_sorted_by_index_desc)
```

usecols = 특정 칼럼들만 가져와라(형식은 list)

```python
# 특정 컬럼만 불러오기

df = pd.read_csv("people_data.csv", usecols=["First Name", "Email"])
df
```

저건 인덱스를 firstname이라고 정해서 출력하는것

```python
# 특정 컬럼을 인덱스로 지정

df = pd.read_csv("people_data.csv", index_col="First Name")
print(df)
```

N/A나 unknown값이 있다면 저건 NAN 로 바꿔라 라고 명령, scv파일에서 어떤게 결측치인지는 아무도 모르기 떄문어 저렇게 여러 조건을 걸음

```python
# 이런 값이 있으면 NaN으로 바꾸어라 → 결측치 처리

df = pd.read_csv("people_data.csv", na_values=["N/A", "Unknown"])
df
```

nrows = 3 세줄만 불러와라

```python
# 몇 라인만 불러오기

df = pd.read_csv('people_data.csv', nrows=3)
df
```

```python
# 첫 몇 라인은 건너뛰기

df = pd.read_csv("people_data.csv", skiprows = [4,5])
print("Dataset After skipping rows: ")
df
```
