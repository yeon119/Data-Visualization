```python
# list comprehension
#이건 파이썬 내용
# 1. 기존 리스트의 모든 요소에 특정 연산을 적용하여 새 리스트를 만드는 경우
squares = [x**2 for x in range(10)] #[■ for ■ in ◆]
print(squares)

words = ['python', 'is', 'awesome']
upper_words = [w.upper() for w in words]
print(upper_words)

# 2. 조건에 맞는 요소만 골라내고 나머지는 버림
evens = [x for x in range(1, 21) if x % 2 == 0] #[■ for ■ in ◆ if ●]
print(evens)

fruits = ['apple', 'banana', 'cherry', 'kiwi', 'mango']
a_fruits = [f for f in fruits if 'a' in f]
print(a_fruits)

# 3. 조건에 의해서 새로운 리스트 만들기
numbers = [1, 2, 3, 4, 5]
labels = ['Even' if x % 2 == 0 else 'Odd' for x in numbers] #[■ if ● for ■ in ◆]
print(labels)

# 4. for안에 또 다른 for
matrix = [[1, 2], [3, 4], [5, 6]]
flat = [num for row in matrix for num in row]

print(flat)
```

```python
matrix = [[1, 2], [3, 4], [5, 6]]

for row in matrix: #[1, 2], [3, 4], [5, 6]
    for num in row: #1,2,3,4,5,6
        print(num)

# 두 for를 한 줄에 쓰면
# for row in matrix for num in row

# 마지막 num을 앞쪽에
# num for row in matrix for num in row
```

```python
# 5. dictionary comprehension

square_dict = {x: x**2 for x in range(5)}
print(square_dict)
```

```python
# 정리

import pandas as pd

data = {
    'state': ['California', 'New York', 'Texas', 'Florida'],
    'year': [2000, 2010, 2000, 2010],
    'pop': [33.8, 37.2, 20.8, 18.8]
}
df = pd.DataFrame(data)
df
```

```python
# 1. df[]는 기본적으로 컬럼 선택 → 컬럼 타입은 대체로 문자열
# 데이터 프레임이라 하고 []을 쓰면 컬럼 선택임 
col_state = df['state'] # 여기까지는 컬럼 하나 선택
print(col_state)

# 리스트를 넣어서 여러 컬럼 선택
cols = df[['state', 'pop']] # 얘는 여러개의 칼럼 선택  []가 메인임
print(cols)
```

```python
# 2. df[]에 :를 사용하면 행 선택

rows = df[0:2] # 대괄호에 콜론을쓰면 행을 선택 0~1까지
print(rows)

# df[]에 쉼표가 없다
# 1. 슬라이스 → 행 선택
# 2.   컬럼명 → 열 선택, [컬럼1, 컬럼 2, ...]은 쉼표가 아니라 단일 리스트를 넘긴 것
```

```python
# 3. 불리언 마스크를 사용하면 행 선택

mask = df['pop'] > 30
print(df[mask])
```

```python
# 컬럼 이름을 알려주지 않음

df = pd.DataFrame([[10, 20], [30, 40]])
# pandas가 디폴트 컬럼 이름 0, 1, 2, ...을 사용
# 이 경우 컬럼명은 숫자 타입에 해당
print(df)
print('-'*100)
print(df[0]) 

#[]에 뭔가를 쓰면 컬럼 : 을쓰면 행 df[0] => 은 컬럼을 선택
#숫자를 썼는데 컬럼인 이유? => 컬럼이 0이어서?
#loc, iloc = > 괄호를 쓰는데 loc는 인덱스를 그대로 iloc는 몇번쨰인지 정함 :를 쓰면 행, 열로 나누어짐
# print(df['0']) # KeyError: '0'

print('-'*100)
print(df[0:2])
```

```python
# 인덱스 이름 - 다 만들고 나중에 이름 붙이기

rng = np.random.RandomState(42)
rng.rand(6,3)
```

```python
# rng.rand(6, 3)는 난수 생성기(rng)를 사용해서 6행 3열의 2차원 배열을 생성
df = pd.DataFrame(rng.rand(6, 3),  #6행 3열
                  index=[['A', 'A', 'B', 'B', 'C', 'C'], [1,2,1,2,1,2]],
                  columns= ['col1', 'col2', 'col3']
                  )

df
```

```python
df.index.names = ['index1', 'index2'] #인덱스 이름을 붙임
df
```

```python
# 인덱스 이름 - 만들면서 알려주기

index_arr = [['A', 'A', 'B', 'B', 'C', 'C'], [1,2,1,2,1,2]]
index = pd.MultiIndex.from_arrays(index_arr, names= ('index1', 'index2'))# name은 name인데 인덱스의 name임 저걸 바꾸면 인덱스 이름이 바뀜
#                                            ↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑

df = pd.DataFrame(rng.rand(6,3), 
                  index=index,
                  columns= ['col1', 'col2', 'col3']
                  )

df
```
