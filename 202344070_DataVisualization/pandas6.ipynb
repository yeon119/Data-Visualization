```python
import pandas as pd

# 멀티인덱스 컬럼을 가진 DataFrame
df = pd.DataFrame(
    [[80, 85, 90, 95],
     [70, 75, 78, 82]],
    index=['A', 'B'], # 인덱스 지정
    columns=pd.MultiIndex.from_tuples([ # 칼럼 이름을 지정 튜플이 4개 있는 리스트
        ('1학기', '국어'), ('1학기', '수학'),
        ('2학기', '국어'), ('2학기', '수학') # 각 튜플의 첫번쨰는 레벨0 두번쨰는 레벨 1 
    ])
)
df
# 1학기 2학기에 해당하는 부분이 레벨0
# 국어 수학은 레벨 1
# 즉 레벨 2의 칼럼임
```

```python
df=pd.DataFrame(
    {"key1":list("aabbab"),
     "key2":["one","two","three"]*2,
     "data1":np.random.randn(6),
     "data2":np.random.randn(6)})
df
```

```python
df["data1"] # 
```

```python
group1=df["data1"].groupby(df["key1"])
# 그룹바이 왼쪽은 대상 = data1 // key1은 기준
# key1 컬럼의 값을 기준으로         ←-------- 기준: 컬럼
# data1 컬럼을                    ←-------- 대상: Series
# 그룹으로 나누어라
```

```python
df[["key1", "data1"]]
#기준에 해당하는 컬럼과 대상을 지정
```

```python
# group1=df["data1"].groupby(df["key1"])에서 group은

# "기준이 무엇인 시리즈" ←----- 그룹 1
# "기준이 무엇인 시리즈" ←----- 그룹 2
# ...
# "기준이 무엇인 시리즈" ←----- 그룹 n
# 로 구성

# group1
# ├── 'a'             ←----- 그룹명
# │    0   -0.2816
# │    1   -0.3085
# │    4   -2.6236
# │
# └── 'b'             ←----- 그룹명
#      2    1.6443
#      3   -0.6814
#      5   -0.5932

group1 # 그룹만 실행하면 아무것도 출력되지 않음
```

```python
# 1. 그룹 목록 확인
group1.groups

# {'a': [0, 1, 4], 'b': [2, 3, 5]}  a라 해당하는 인덱스는 0,1,4 b라 해당하는 인덱스는 2,3,5
# 키: 그룹명
# 값: 해당 그룹의 인덱스 리스트
```

```python
# 2. 각 그룹 순회하며 내용 출력

# group
# ├── 'a'             ←----- 그룹명 (name)
# │    0   -0.2816
# │    1   -0.3085
# │    4   -2.6236
# │
# └── 'b'             ←----- 그룹명 (name)
#      2    1.6443
#      3   -0.6814
#      5   -0.5932

for name, _ in group1: # 이름만 출력함
    print(name)

print('-'*100)

for _, g in group1: # 그룸의 누가 있는가를 출력 
    print(g)# 시리즈를 두개 출력하는데 a랑 b를 출력

print('-'*100)

for name, g in group1: #이름과 시리즈를 모두 출력함
    print(name, g) 
```

```python
# 3. 특정 그룹
group1.get_group("a") #특정한 그룹만 출력 get_group
```

```python
group1.mean() # 집계함수 집계란? 글자가 있고 말이 있음  mean = 평균 
# key1은 그럼이 이렇게 나누어져있다
# a의 평균 , b의 평균 을 알려줌
```

```python
group2=df["data1"].groupby([df["key1"], df["key2"]]) #key값이 두개
# key1 컬럼, key1 컬럼의 값을 기준으로         ←-------- 기준: 컬럼
# data1 컬럼을                              ←-------- 대상: Series
# 그룹으로 나누어라 → key1, key2가 모두 같아야 같은 그룹
```

```python
df[["key1", "key2", "data1"]] # k1,k2는 기준 data1은 대상 
#a1,a2 b1,b3
```

```python
for name, _ in group2:
    print(name) # 그래서 이름이 두개

print('-'*100)

for _, g in group2:
    print(g)

print('-'*100)

for name, g in group2:
    print(name, g)
```

```python
ave = group2.mean()
ave # mean은 평균을 의미하고 a의1,2 b의 1,3의 평균을 각각 표현함
# 타입이 멀티 인덱스
```

```python
ave.index

# MultiIndex([('a',   'one'),
#             ('a',   'two'),
#             ('b',   'one'),
#             ('b', 'three')],
#            names=['key1', 'key2']) #key1 레벨0에 붙은 인덱스 이름 key2는 레벨 1에 붙은 인덱스 이름

# 레벨 0: key1 ('a', 'b')  ← 바깥쪽
# 레벨 1: key2 ('one', 'two', 'three') ← 안쪽
```

```python
# unstack()은 인덱스를 컬럼으로 이동한다는 점에서 reset_index()와 유사
# 하지만 unstack()은 멀티인덱스 상황에서 특정 인덱스를 컬럼으로 이동

# unstack()의 디폴트는 가장 마지막 레벨을 컬럼으로
# 즉, 디폴트는 unstack(-1) -1은 마자막 레벨을 의미

ave.unstack()
# 결측치가 생기는 이유는 {a, b} x {one, two , three} 모든 조합이 있지 않기 때문
# 결측치는 a에는 three가 없으니 생기고 b에는 two가 없으니 생김
```

```python
ave.unstack(0)
# 평균에서 결측치가 생기는 이유도 위와같음
```

```python
df3 = pd.DataFrame({
    "팀": ["A", "A", "B", "B", "A", "B"],
    "점수": [85, 90, 78, 92, 88, 76],
    "출석": [20, 18, 22, 19, 21, 20]
})

group4 = df3.groupby("팀")
group4.mean()

# 나중에 밑에 있는 기본 통계 함수 부분 다시 넣어서 공부
```

```python
# 그룹 내의 여러 값을 하나로 줄여주는 함수를 집계 함수라고 함
# [0, 3]  →   sum(): 3
# [0, 3]  →  mean(): 1.5
# [0, 3]  →   min(): 0
# [0, 3]  →   max(): 3

# 기본 통계
# 'sum'     # 합계
# 'mean'    # 평균
# 'median'  # 중앙값
# 'min'     # 최솟값
# 'max'     # 최댓값
# 'std'     # 표준편차
# 'var'     # 분산
# 'count'   # 개수 (NaN 제외)
# 'size'    # 개수 (NaN 포함)

# 기타
# 'first'   # 첫 번째 값
# 'last'    # 마지막 값
# 'nunique' # 고유값 개수
```

```python
rng = np.random.RandomState(0)
df = pd.DataFrame({
    'key': ['A', 'B', 'C', 'A', 'B', 'C'],
    'data1': range(6),
    'data2': rng.randint(0, 10, 6)
})
df
```

```python
# aggregate()는 여러 집계 함수를 한번에 적용
df.groupby('key').aggregate(['min', np.median, 'max']) #min max는 문자열 처럼 사용 np.median 은 넘파이에서 지원하는 형식
```

```python
# groupby('key')는 컬럼 'key'의 unique 값을 인덱스로
# key칼럼에 있는 값을 기준으로 표준편차를 계산하라
df.groupby('key').std()
```

```python
#lambda
#lambda함수 -> 1회용 함수 
#그자리에서 만들어 쓰고 끝 
lambda x:x #앞에 x는 내가 넘겨주는 인자값 =  parameter :뒤에는 함수 내용
#lambda parameter:function
lambda x:x+10 #=> 입력을 받아서 10을 더해 리턴을 한다

x = lambda a,b : a*b
print(x(5*6)) # => 30

# 이름이 없음 아무개 함수?
#
```

```python
df.groupby('key').filter(lambda x: x['data2'].std()>4) 

# filter(조건)의 형태로 사용

# lambda x: 에서 x는 그룹에 해당; 그룹의 타입은 데이터프레임
# lambda x: x['data']는 그룹의 'data' 컬럼에 해당

# "data2 컬럼의 표준편차가 4보다 큰 그룹만 남겨라"
# 람다함수 나중에 다시 복습
```

```python
df.groupby('key').filter(lambda x: len(x) == 2)        # 그룹 행 개수가 2 초과
# 약간 이제 조건을 거는 느낌으로 길이가 2인 
```

```python
df.groupby('key').filter(lambda x: x['data2'].max() > 8)
```

```python
# use map with lambda function
fruits = ['apple','banana','cherry'] #문자열이 들어있는 리스트
lengths =list(map(lambda x: len(x), fruits)) # map은 람다를 여러곳에서 하고싶을떄 사용
print(lengths)
```

```python
# Use filter with ;ambda function
numbers = [1,2,3,4,5,6,7,8]
evens = filter(lambda x:x %2 ==0 ,numbers) # 얘는 2로 나눈값이 0인지를 확인하는거ㅗㄴ데 true false로 나오고 그거에 대한 값을반환
print(list(evens))
```
